```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])

print("==== Zomato sales analysis ==== ")
print("\n Sales data shape", sales_data.shape)
print("\n Sample data for 1st 3 resturent: ", sales_data[:3])
print("\n Sample data for 1st 3 resturent: ", sales_data[:,1:])
```

```python
cumsum=np.cumsum(sales_data[:,1:],axis=1)
print("\n Cumulative sum of sales per resturent:", cumsum)
```

```python
plt.figure(figsize=(8, 6))
plt.plot(np.mean(cumsum, axis=0))
plt.title("Average Cumulative Sales Over Years")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
         
```

```python
vector1=np.array([1, 2, 3,4,5])
vector2=np.array([6, 7, 8,9,10])

print("\n Vector1: ", vector1+vector2)

print("\n Multiplication ",(vector1*vector2))

print("\n Dot product:",np.dot(vector1,vector2))

angle = np.arccos(np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2)))

print("total angle: ",angle)
```

```python
resturent_types=np.array(['biryani','chinese','pizza','burger','cafe'])

# Create a vectorized function to convert strings to uppercase
# Uppercase means converting all characters in a string to their capital form.
vectorized_upper = np.vectorize(str.upper)
print("vectorized upper",vectorized_upper(resturent_types))
```
